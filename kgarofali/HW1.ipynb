```python
import numpy as np

data = np.genfromtxt('asteroids5000.csv', delimiter=',')
mask = np.isnan(data)
print "There are %d values missing from the data" % len(data[mask])

bad = np.where(np.isnan(data))
data_nonan = np.delete(data, bad[0], axis=0)
print "There are %d complete rows in the data set" % data_nonan.shape[0]

# just to check that it pulled out the correct number, remove duplicate row numbers from 'bad'
tmp = len(np.array(list(set(bad[0]))))
if data_nonan.shape[0] != (data.shape[0] - tmp):
    print "Oops!"
else: 
    pass

# now for min, max, mean and std in each column
def compute_stuff(x):
    mins = np.zeros(x.shape[1])
    maxes = np.zeros(x.shape[1])
    means = np.zeros(x.shape[1])
    stds = np.zeros(x.shape[1])
    for i in range(x.shape[1]):
        mins[i] = np.min(x[:,i])
        maxes[i] = np.max(x[:,i])
        means[i] = np.mean(x[:,i])
        stds[i] = np.std(x[:,i])
    return mins, maxes, means, stds
compute_stuff(data_nonan) # gives arrays of mins, maxes, means, and stds for each column. 

```

The above cell should print out an array that gives the mins, maxes, means and standard deviations for each column. 

```python
%pylab inline 
import matplotlib.pyplot as plt
!head -1 asteroids5000.csv

fig = plt.figure()
plt.scatter(data_nonan[:,0],np.sin(np.radians(data_nonan[:,2])))
plt.xlabel('Semi-Major Axis [AU]')
plt.ylabel('Sin(i)')

```

There seem to be discrete distributions of asteroids in terms of the semi-major axes. The gaps exist due to resonances that the asteroid belt has with Jupiter--asteroids are usually kicked out of the orbits in which there is a strong resonance. 

```python
fig = plt.figure()
plt.scatter(data_nonan[:,3], data_nonan[:,5])
plt.plot([2, 14], [0.78,0.78], color="red")
plt.ylabel('B-V')
plt.xlabel('H')
```

```python
fam1 = np.where(data_nonan[:,5] <= 0.78)
fam2 = np.where(data_nonan[:,5] > 0.78)
fig = plt.figure()
plt.scatter(data_nonan[:,0][fam1],np.sin(np.radians(data_nonan[:,2][fam1])), color="green")
plt.scatter(data_nonan[:,0][fam2],np.sin(np.radians(data_nonan[:,2][fam2])), color="cyan")
plt.xlabel('Semi-Major Axis')
plt.ylabel('Sin(i)')
```

The family with the larger B-V mag seems to have the smaller semi-major axes, while the families with the smaller B-V mags span the whole range of semi-major axes. This is roughly in line with what Parker et al. found--i.e. different magnitudes corresponding to different families with different orbital characteristics.
