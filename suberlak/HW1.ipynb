# Part 1: loading and exploring the data

#### 1) Load the data

```python
file='/Users/astronomy/Astr599/2013_fall_ASTR599/notebooks/data/asteroids5000.csv'
aster=np.genfromtxt(file,delimiter=',')
```

#### 2) Find the missing values: calculate their number; and calculate the number of complete rows

```python
print "Number of values ", aster.size
missing=np.isnan(aster)
print "Number of missing entries (NaN): ", missing.sum()
print "Number of complete rows",np.all(~missing,1).sum()
```

#### 3) New array with no missing values

```python
goodaster = aster[np.all(~missing,1),:]
print "Check: number of missing values ", np.isnan(goodaster).sum()
print "New array number of rows (as above)", goodaster.shape[0]
```

#### 4) Table of maximum, minimum, mean, and standard deviation of the values in each column.

```python
goodaster.shape
print 'Max: %6.3f %6.3f %6.3f %6.3f %6.3f %6.3f %6.3f' % tuple(goodaster.max(0).reshape(-1,).tolist())
print 'Min: %6.3f %6.3f %6.3f %6.3f %6.3f %6.3f %6.3f' % tuple(goodaster.min(0).reshape(-1,).tolist())
print 'Mean: %6.3f %6.3f %6.3f %6.3f %6.3f %6.3f %6.3f' % tuple(goodaster.mean(0).reshape(-1,).tolist())
print 'S Dev: %6.3f %6.3f %6.3f %6.3f %6.3f %6.3f %6.3f' % tuple(goodaster.std(0).reshape(-1,).tolist())
```

I notice in that ditribution that there are certain values of the semimajor axis which are not populated by asteroids (eg. around 3.3, 3.7 au). Those gaps could possibly be explained by the orbital resonances (http://hpcc.astro.washington.edu/stawarz/orbres.html) 

#### 4) Plot of a color-magnitude diagram of the asteroids (H vs B-V) (overplotted a line dividing the distributions)

```python
# numbers of relevant columns of goodaster : 
#   0 'a', 1 'e', 2 'i' , 3 'H', 4 'diameter', 5 'BV', 6 'UB' 

plt.scatter(goodaster[:,5],goodaster[:,3])   # BV vs H 
plt.ylabel("H")
plt.xlabel("B-V")
plt.ylim(14,2)
bv_lim=0.78
plt.vlines(bv_lim,*plt.ylim())
```

#### 5) Plot of a color-magnitude diagram of the asteroids, in different colours

```python
green=(goodaster[:,5] < bv_lim)  # search for indices fulfilling that criterion
magenta=(goodaster[:,5] >= bv_lim)

plt.scatter(goodaster[green,0],np.sin(np.deg2rad(goodaster[green,2])),c='g',label="B-V < 0.78")  # 0 is 'a', 2 is 'i'
plt.scatter(goodaster[magenta,0],np.sin(np.deg2rad(goodaster[magenta,2])),c='m',label="B-V >= 0.78")
plt.xlabel("Semimajor Axis")
plt.ylabel("sin (inclination angle)")
plt.legend()
```

It seems that the orbit of magenta  asteroids is kept below a=3.5 au , whereas for green asteroids this is not the case. This indicates that the color-based selection (of those two families from B-V vs H diagram) represents some true physical property of those families (an existence of a correllation). At  a=2.5 one of the Kirkwood gaps is visible.

#### 6) Comparison to Parker et al. 2008 data

Upon inspection of Fig.4 in Parker et al. 2008 we see that our magenta coloured points would correspond to stony (S-type) asteroids, while green correspond to carbonaceous (C-type) asteroids.
